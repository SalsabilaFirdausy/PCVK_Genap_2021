```
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

print('Mengubah tingkat kecerahan citra')
print('--------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/week3-fish.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

#Akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

#Cara simple tanpa loop
#brightness_image = cv.convertScaleAbs(original, beta=brightness)

final_frame = cv.hconcat((original,brightness_image))
cv2_imshow(final_frame)
```

```
#Assignment 1

print('Mengubah kontras dan tingkat kecerahan citra')
print('--------------------------------')
try :
  brightness = int(input('Masukkan tingkat kecerahan [-255 - 255]: '))
  contrast = float(input('Masukkan kontras [1.0 - 3.0]: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/week3-woman.JPG')
contrast_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      contrast_image[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)


final_frame = cv.hconcat((original, contrast_image))
cv2_imshow(final_frame)
```

```
#Assignment 2

print('Mengubah tingkat kecerahan citra dengan Transformasi')
print('--------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/week3-woman.JPG')
brightness_image = np.zeros(original.shape, original.dtype)

#Akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness*(np.log(1+original[y,x,c])),0,255)

final_frame = cv.hconcat((original,brightness_image))
cv2_imshow(final_frame)
```

```
#Assignment 3 - Averaging
img = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_average = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img_average = cv.cvtColor(img_average,cv.COLOR_BGR2RGB)

#image[x,y,c]
R, G, B = img[:,:,0], img[:,:,1],img[:,:,2]

average = R/3 + G/3 + B/3
img_average[:,:,0] = average
img_average[:,:,1] = average
img_average[:,:,2] = average
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img)
axs[1].imshow(img_average)
```

```
#Assignment 3 - Lightness
img = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_lightness = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img_lightness = cv.cvtColor(img_lightness,cv.COLOR_BGR2RGB)

#Formula
R, G, B = img_lightness[:,:,0], img_lightness[:,:,1], img_lightness[:,:,2]
lightness = np.maximum (R,G,B) / 2 + np.minimum(R,G,B) / 2

img_lightness[:, :, 0] = lightness
img_lightness[:, :, 1] = lightness
img_lightness[:, :, 2] = lightness
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img)
axs[1].imshow(img_lightness)
```

```
#Assignment 3 - Luminance
img = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_lightness = cv.imread('/content/drive/MyDrive/PCVK/week3-baboon.jpg')
img_lightness = cv.cvtColor(img_lightness,cv.COLOR_BGR2RGB)

#Formula
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
luminance = 0.21 * R + 0.72 * G + 0.07 * B

img_lightness[:, :, 0] = luminance
img_lightness[:, :, 1] = luminance
img_lightness[:, :, 2] = luminance
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img)
axs[1].imshow(img_lightness)
```
